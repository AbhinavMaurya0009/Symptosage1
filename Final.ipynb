## Feature Engineering

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import mutual_info_classif, f_classif
from sklearn.linear_model import LassoCV
import numpy as np

# Initialize Random Forest Classifier
rf_classifier = RandomForestClassifier()

# Initialize RFE with Random Forest Classifier
rfe_selector = RFE(estimator=rf_classifier, n_features_to_select=25, step=1)

# Fit RFE
rfe_selector.fit(X, y_encoded)

# Get selected features
selected_features_rfe = set(X.columns[rfe_selector.support_])

# Calculate Mutual Information
mutual_info_scores = mutual_info_classif(X, y_encoded)

# Calculate Information Gain (using f_classif)
information_gain_scores = f_classif(X, y_encoded)[0]

# Initialize LASSO model
lasso_selector = LassoCV(cv=5)

# Fit LASSO
lasso_selector.fit(X, y_encoded)

# Get selected features
lasso_coefs = np.abs(lasso_selector.coef_)
selected_features_lasso = set(X.columns[lasso_coefs != 0])

# Define threshold for Mutual Information and Information Gain scores
threshold = 0.1  # Adjust this threshold as needed

# Get selected features based on Mutual Information and Information Gain scores
mutual_info_selected_features = set(X.columns[mutual_info_scores > threshold])
information_gain_selected_features = set(X.columns[information_gain_scores > threshold])

# Find common features among all feature selectors
selected_features_common = selected_features_rfe.intersection(selected_features_lasso, mutual_info_selected_features, information_gain_selected_features)

# Limit to 25 features if there are more than 25 common features
selected_features_common = list(selected_features_common)[:25]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Define Random Forest Classifier
rf_classifier = RandomForestClassifier()

# Define hyperparameters grid for grid search
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_leaf': [1, 2, 4]
}

# Implement Grid Search
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5, scoring='f1_micro', n_jobs=-1)
grid_search.fit(X_selected, y_encoded)

# Obtain best estimator
best_rf_classifier = grid_search.best_estimator_
```

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer, precision_score, recall_score, f1_score, hamming_loss
from sklearn.model_selection import KFold

# Define the number of splits for cross-validation
num_splits = 5  # You can adjust this as needed

# Define the KFold cross-validation splitting strategy
k_fold = KFold(n_splits=num_splits, shuffle=True, random_state=42)

from sklearn.metrics import make_scorer

# Define scoring metrics
scoring = {
    'precision_micro': make_scorer(precision_score, average='micro'),
    'recall_micro': make_scorer(recall_score, average='micro'),
    'f1_micro': make_scorer(f1_score, average='micro'),
    'hamming_loss': make_scorer(hamming_loss)
}

# Perform cross-validation using cross_validate
cv_results = cross_validate(best_rf_classifier, X_selected, y_encoded, cv=k_fold, scoring=scoring)

# Average performance metrics across folds
precision_micro_avg = cv_results['test_precision_micro'].mean()
recall_micro_avg = cv_results['test_recall_micro'].mean()
f1_micro_avg = cv_results['test_f1_micro'].mean()
hamming_loss_avg = cv_results['test_hamming_loss'].mean()

print("Average Precision (micro):", precision_micro_avg)
print("Average Recall (micro):", recall_micro_avg)
print("Average F1-score (micro):", f1_micro_avg)
print("Average Hamming Loss:", hamming_loss_avg)
```
